#### Read the dataset (df= pd.read_csv(r'………………………….\Titanic.csv')
- Display the first and last 10 instances from the dataset
- Acquire the necessary information using the df.info() and df. Describe().
- Retrieve the number of columns and rows. (using shape)


```python
import pandas as pd

# READING FROM THE DATASET
__dataSet__ = pd.read_csv(r'Titanic.csv')

# DISPLAYING THE FIRST 10 INSTANCES
print("FIRST 10 INSTANCES :")
print(__dataSet__.head(10))

# DISPLAYING THE LAST 10 INSTANCES
print("\nLAST 10 INSTANCES :")
print(__dataSet__.tail(10))

# FILTERING THE REQUIRED INFROMATION
print("\nDATAFRAME INSTANCES :")
__dataSet__.info()

# STATISTICS
print("\nDATAFRAME DESCRIPTION :")
__dataSet__.describe()

# EXTRACTING THE ROWS AND COLUMNS
num_rows, num_columns = __dataSet__.shape
print(f"\nROWS : {num_rows}")
print(f"COLUMNS : {num_columns}")
```

#### Create the data visualization using the matplotlib.
- Visualize the Gender of Passengers using the Bar graph.
- Visualize the Survival Count of Passengers using the Bar graph.
- Visualize the Age of Passengers using the Bar/Histogram graph.
- Visualize the comparison of Age and Fare of Passengers using the Scatterplot.


```python
import pandas as pd
import matplotlib.pyplot as plt



# VISUALIZATION OF THE GENDER
gender_counts = __dataSet__['Sex'].value_counts()
plt.figure(figsize=(15, 6))
plt.bar(gender_counts.index, gender_counts.values, color = 'Yellow')
plt.title('GENDER OF THE PASSENGER')
plt.xlabel('GENDER')
plt.ylabel('COUNT')
plt.show()

# SURVIVAL COUNT
survival_counts = __dataSet__['Survived'].value_counts()
plt.figure(figsize=(15, 6))
plt.bar(survival_counts.index, survival_counts.values, color = 'Red')
plt.title('SURVIVAL COUNT OF THE PASSENGERS')
plt.xlabel('SURVIVED')
plt.ylabel('COUNT')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()

# AGE OF THE PASSENGER
plt.figure(figsize=(15, 6))
plt.hist(__dataSet__['Age'].dropna(), bins=20, edgecolor='k', alpha=0.7)
plt.title('AGE DISTRIBUTION OF THE PASSENGERS')
plt.xlabel('AGE')
plt.ylabel('COUNT')
plt.show()

# AGE AND FAIR
plt.figure(figsize=(15, 6))
plt.scatter(__dataSet__['Age'], __dataSet__['Fare'], alpha=0.5)
plt.title('AGE V FARE OF THE PASSENGERS')
plt.xlabel('AGE')
plt.ylabel('FARE')
plt.show()
```
